# for

```python
'''
구구단

단수    x   곱하는 수
(2-9)      (1-9)
2          (1-9)
3          (1-9)
4          (1-9)
5          (1-9)
6          (1-9)
7          (1-9)
8          (1-9)
9          (1-9)
'''

for num in range(1, 10) :
    print('2 x', num, '=', 2 * num)
```

```python
for dan in range(1, 10) :
    for num in range(1, 10) :
        print(dan, 'x', num, '=', dan * num)
    print('-' * 20)
```

```python
for idx in range(1, 4) :
    for dan in range(1, 4) :
        for num in range(1, 4) :
            print(idx, '-', dan, '-', num)
        print('-' * 20)
```

```python
for i in range(2, 10) :
    print(f"-- {i} 단 --")
    for j in range(1, 10) :
        print(f"{i} x {j} = {i * j}")
    print()
```
