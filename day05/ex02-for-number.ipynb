# for

```python
'''
구구단의 홀수 단만 출력하는
for 문을 작성하세요
'''

for i in range(2, 10) :
    if i % 2 == 1 :
        for j in range(1, 10) :
            print(f"{i} x {j} = {i * j}")
        print()
```

```python
'''
구구단의 짝수 단만 출력하는
for 문을 작성하세요
'''

for i in range(2, 10) :
    if i % 2 == 0 :
        for j in range(1, 10) :
            print(f"{i} x {j} = {i * j}")
        print()
```

```python
'''
구구단의 각 단에 홀수만 곱해서
출력하는 for 문을 작성하세요
'''

for i in range(2, 10) :
    for j in range(1, 10) :
        if j % 2 == 1 :
            print(f"{i} x {j} = {i * j}")
    print()
    
```

```python
'''
짝수단에는 짝수만 곱하고
홀수단에는 홀수만 곱하는
구구단을 출력하는
for 문을 작성하세요
'''

for i in range(2, 10) :
    for j in range(1, 10) :
        if i % 2 == 0 and j % 2 == 0 :
            print(f"{i} x {j} = {i * j}")
        elif i % 2 == 1 and j % 2 == 1 :
            print(f"{i} x {j} = {i * j}")
        
    print()
    
```
