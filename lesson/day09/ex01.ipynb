```python
'''
108 page
'''

# 3. 검색
dataset = [5, 10, 18, 22, 35, 55, 75, 103]
value = int(input("검색할 값 입력 : "))

low = 0
high = len(dataset) - 1
loc = 0
state = False

while (low <= high):
    mid = (low + high) // 2

    if dataset[mid] > value:
        high = mid - 1
    elif dataset[mid] < value:
        low = mid + 1
    else: # 찾은 경우
        loc =  mid
        state = True
        break # 반복 exit

if state :
    print('찾은 위치 : %d 번째' %(loc+1))
else :
    print('찾는 값은 없습니다.')
```

```python
'''
127 page
'''

# 3) 몬테카를로 시뮬레이션
import random

# 단계 1 : 동전 앞면과 뒷면의 난수 확률분포 함수 정의
def coin(n) :
    result = []
    for i in range(n) :
        r = random.randint(0, 1)
        if (r == 1) :
            result.append(1) # 앞면
        else :
            result.append(0) # 뒷면
    return result

print(coin(10)) 
print("=" * 50)
```

```python
# 단계 2 : 몬테카를로 시뮬레이션 함수 정의
def montaCoin(n) :
    cnt = 0
    for i in range(n) :
        cnt += coin(1)[0] # coin 함수 호출

    result = cnt / n # 동전 앞면과 뒷면의 누적 결과를 시행 횟수(n)로 나눈다.
    return result
```

```python
#단계 3 : 몬테카를로 시뮬레이션 함수 호출
print(montaCoin(10)) # 0.3
print(montaCoin(30)) # 0.5666667
print(montaCoin(100)) # 0.53
print(montaCoin(1000)) #  0.51
print(montaCoin(10000)) # 0.5033
```

```python
'''
142 page
'''
def Adder(n):
    if n == 1:  # n=1 -> exit 조건
        return 1
    else:
        # 재귀호출
        result = n + Adder(n - 1)
        # 5(first)->4(5-1) -> 3(4-1) -> 2(3-1) | [1(2-1)]
        '''
        1. stack(5, 4, 3, 2)
        2. 덧셈 : 1 + [2 + 3 + 4 + 5 ]         
        '''
        print(n, end=' ')  # 2 3 4 5
        return result  # 15


print('n=1 :', Adder(1))  # n=1 : 1
print('\nn=5 :', Adder(5))  # n=5 : 15
```

```python
try:
    # db 연결하기 - Connection 객체 생성하기
    # sqlite3 묘듈에 정의 되어있는 connect() 함수를 호출함
    conn = sqlite3.connect("C:/Users/user/python_week/test/day09/data/sqlite_db")
    # sql 실행 객체 생성하기
    cursor = conn.cursor()

    # table 생성하기
    sql = "create table if not exists test_table (name text(10), phone text(15), addr text(50))"

    # sql 문 실행하기
    cursor.execute(sql)

    # record 추가하기
    cursor.execute("insert into test_table values('이순신','010-1234-1234','서울시')")
    cursor.execute("insert into test_table values('더조은','010-2222-2222','광주시')")
    cursor.execute("insert into test_table values('아이티','010-3333-3333','부산시')")

    # DML 작업을 최종적으로 DB 에 반영함
    conn.commit()

    # 레코드 조회하기 : * (모든 컬럼을 의미함)
    cursor.execute("select * from test_table")
    rows = cursor.fetchall()

    for row in rows:
        print(row)
    print("-" * 50)    

    print("이름 \t전화번호 \t\t주소")
    for row in rows:
        print(row[0],'\t',row[1],'\t',row[2])

except Exception as e:
    print("DB 연결 실패 :",e)
    # DML 작업한 부분 실행 취소
    conn.rollback()
    
finally:
    cursor.close()
    conn.close()
```
