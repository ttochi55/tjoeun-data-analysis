```python
file1 = open('../dataset/newfile.txt', 'a', encoding='utf-8')

for idx in range(11, 21) :
    data = f"{idx} 번째 줄(행) 입니다.\n"
    file1.write(data)
    
file1.close()
```

```python
file2 = open('../dataset/newfile.txt', 'r', encoding='utf-8')

while True:
    data = file2.readlines()
    if not data:
        break
    for str_line in data:
        print(str_line, end="")
        
file2.close()
```

```python
with open('../dataset/info.txt', 'r', encoding='utf-8') as file5:
    for line in file5:
        name, weight, height = line.strip().split(', ')
        
        if (not name) or (not weight) or (not weight):
            continue
        
        # 체질량 수 구하기
        bmi = int(weight) / ((int(height) / 100) ** 2)
        
        # 정상: 18.5 <= bmi <= 24
        result = ""
        
        if 25 <= bmi:
            result = '과체중'
        elif 18.5 <= bmi:
            result = '정상제중'
        else:
            result = '저체중'
            
        print('\n'.join([f"이름: {name}", f"몸무게: {weight}", f"키: {height}", f"BMI: {bmi}", f"결과: {result}"]))
        print('-' * 25)
        
```
