### DataFrame, list, dictionary, ndarray 상호 변환하기

* list, ndarray 에서 DataFrame 으로 변환하기

```python
import numpy as np
```

```python
col_name1 = ['col1']
list1 = [1, 2, 3]
array1 = np.array(list1)

print('-- 1차원 리스트로 생성한 DataFrame --')
list1_df = pd.DataFrame(list1, columns=col_name1)
print(list1_df.head())
print()

print('-- 1차원 ndarray로 생성한 DataFrame --')
array1_df = pd.DataFrame(array1, columns=col_name1)
print(array1_df.head())
```

```python
# column 명 3 개로 구성된 list 생성하기
col_name2 = ['col1', 'col2', 'col3']

# 2행 3열 (2 x 3)의 리스트와 ndarray 를 생성한 후
# 이드을 DataFrame 으로 변환하기
list2 = [[1, 2, 3],
         [11, 22, 33]]
array2 = np.array(list2)

print('-- 2차원 리스트로 생성한 DataFrame --')
list2_df = pd.DataFrame(list2, columns=col_name2)
print(list2_df.head())
print()

print('-- 2차원 ndarray로 생성한 DataFrame --')
array2_df = pd.DataFrame(array2, columns=col_name2)
print(array2_df.head())
```

### DataFrame 데이터 삭제하기 - DataFrame 의 drop(컬럼명, axis=0/1) 메소드

* axis 를 지정해서 삭제하기

```python
'''
원본은 변형되지 않고, 변형된(지정한 컬럼이 삭제된) 새로운 DataFrame 을 반환함
'''
titanic_df.drop('Age_0', axis=1)
titanic_df.head(1)
```

```python
'''
titanic_df 를 출력하면 변함이 없고
drop_titanic_df 를 출력해야 결과를 확인할 ㅅ ㅜ있음
'''
drop_titanic_df = titanic_df.drop('Age_0', axis=1)
drop_titanic_df.head(1)
```

### 여러 개의 column 삭제하기

* drop() 메소드 안에 argument 로 삭제할 컬럼명으로 구성된 리스트를 넣어줌
* inplace=True 로 설정하면 원본이 변형됨 : inplace=False 가 default 값임
*  ㄴ 원본을 변형(원본에서 삭제함)되면서 새로운 DataFrame 을 반환하지 않음

```python
'''
inplace=True 로 설정했기 때문에 원본이 변형됨
'''
# 원본을 변형(원본에서 삭제함)되면서 새로운 DataFrame 을 반환하지 않음
# drop_df = titanic_df.drop(['Age_0', 'Age_by_10', 'Family_No'], axis=1, inplace=True)
# drop_df <-- None
# AttributeError: 'NoneType' object has no attribute 'head'
# drop_df.head(1)

titanic_df.drop(['Age_0', 'Age_by_10', 'Family_No'], axis=1, inplace=True)
titanic_df.head(1)
```
